```python
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
from keras.callbacks import EarlyStopping  
```

```python
from keras.models import Model
```

```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
from keras.datasets import mnist
```

```python
def unison_shuffled_copies(a, b):
    assert len(a)==len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]
```

```python
import pandas as pd
def  reshape(a, labels, num_pixels):
    a = a.reshape(a.shape[0], num_pixels).astype('float32')
    a = a/255
    #labels = np_utils.to_categorical(labels)
    labels = pd.get_dummies(pd.Series(labels)).as_matrix()
    return a, labels
```

```python
def squeeze(a):
    l = list(itertools.chain(*a))
    return l
```

```python
seed = 7
np.random.seed(seed)

(X_train, y_train), (X_test, y_test) = mnist.load_data()
num_pixels = X_train.shape[1] * X_train.shape[2]
```

```python
train_mask = np.isin(y_train, list(range(0, 5)))
test_mask = np.isin(y_test, list(range(0, 5)))
```

```python
X_train_m1, y_train_m1 = X_train.copy()[train_mask], y_train.copy()[train_mask]
X_test_m1, y_test_m1 = X_test.copy()[test_mask], y_test.copy()[test_mask]
```

```python
y_test.shape
```

```python
X_train_m1, y_train_m1 = reshape(X_train_m1, y_train_m1, num_pixels)
X_test_m1, y_test_m1 = reshape(X_test_m1, y_test_m1, num_pixels)

num_classes = y_test_m1.shape[1]
```

```python
y_test_m1.shape[1]
```

```python
def model_1():
    model = Sequential()
    model.add(Dense(100, activation='relu', input_dim=num_pixels, kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    #model.add(BatchNormalization())
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  #optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
              optimizer='adam',
              metrics=['accuracy'])
    
    return model
```

```python
def model_2():
    model = Sequential()
    model.add(Dense(100, activation='relu', input_dim=num_pixels, kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(BatchNormalization())
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  #optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
              optimizer='adam',
              metrics=['accuracy'])
    
    return model
```

```python
def model_3():
    model = Sequential()
    model.add(Dense(100, activation='relu', input_dim=num_pixels, kernel_initializer='he_normal'))
    model.add(Dropout(0.5))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dropout(0.5))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dropout(0.5))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    model.add(Dropout(0.5))
    model.add(Dense(100, activation='relu', kernel_initializer='he_normal'))
    #model.add(Dropout(0.5))
    model.add(BatchNormalization())
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  #optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
              optimizer='adam',
              metrics=['accuracy'])
    
    return model
```

```python
#es = EarlyStopping(monitor='val_loss', mode='min', verbose=1)
#es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=50)
#es = EarlyStopping(monitor='val_loss', mode='min', baseline=0.9)
es = EarlyStopping(monitor='val_acc', mode='max', min_delta=0.001, patience=50)
```

## Part 1

1. Build a DNN with five hidden layers of 100 neurons each, He initialization, and the ELU
activation function.
2. Using Adam optimization and early stopping, try training it on MNIST but only on digits 0
to 4, as we will use transfer learning for digits 5 to 9 in the next exercise. You will need a
softmax output layer with five neurons.
3. Tune the hyperparameters using cross-validation and see what precision you can
achieve.

```python
model = model_1()
history = model.fit(X_train_m1, y_train_m1, validation_data=(X_test_m1, y_test_m1), epochs=300, batch_size=100, verbose=2, callbacks=[es])
# Final evaluation of the model
scores = model.evaluate(X_test_m1, y_test_m1, verbose=0)
print("Baseline Error: %.2f%%" % (100-scores[1]*100))
```

```python
print(history.history.keys())
```

```python
plt.figure(1)

#history of accuracy
plt.subplot(211)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
```

```python
plt.figure(1)

#history of loss
plt.subplot(212)  
plt.plot(history.history['loss'])  
plt.plot(history.history['val_loss'])  
plt.title('model loss')  
plt.ylabel('loss')  
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
plt.show()  
```

The above graph shows overfitting.

4. Now try adding Batch Normalization and compare the learning curves: is it converging
faster than before? Does it produce a better model?

```python
model2 = model_2()
history2 = model2.fit(X_train_m1, y_train_m1, validation_data=(X_test_m1, y_test_m1), epochs=300, batch_size=100, verbose=2, callbacks=[es])
# Final evaluation of the model
scores2 = model2.evaluate(X_test_m1, y_test_m1, verbose=0)
print("Baseline Error: %.2f%%" % (100-scores2[1]*100))
```

```python
plt.figure(1)

#history of accuracy
plt.subplot(211)
plt.plot(history2.history['acc'])
plt.plot(history2.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
```

```python
plt.figure(1)

#history of loss
plt.subplot(212)  
plt.plot(history2.history['loss'])  
plt.plot(history2.history['val_loss'])  
plt.title('model loss')  
plt.ylabel('loss')  
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
plt.show()  
```

It takes more epochs to converge for the same network architecture but with the additional Batch Normalization layer inserted before the final FC layer. However the accuracy of the model is better. We still observe overfitting, however the graphs is getting flactuated unstead of converging unlike the 'model 1', so there is more hope.

In the above graphs we do not see neither the overfitting nor underfitting

```python
model3 = model_3()
history3 = model3.fit(X_train_m1, y_train_m1, validation_data=(X_test_m1, y_test_m1), epochs=300, batch_size=100, verbose=2, callbacks=[es])
# Final evaluation of the model
scores3 = model3.evaluate(X_test_m1, y_test_m1, verbose=0)
print("Baseline Error: %.2f%%" % (100-scores[1]*100))
```

```python
plt.figure(1)

#history of accuracy
plt.subplot(211)
plt.plot(history3.history['acc'])
plt.plot(history3.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
```

```python
plt.figure(1)

#history of loss
plt.subplot(212)  
plt.plot(history3.history['loss'])  
plt.plot(history3.history['val_loss'])  
plt.title('model loss')  
plt.ylabel('loss')  
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
plt.show()  
```

With the Dropout Layer added after each FC layer the training accuracy is much lower as opposed to the validation accuracy in the first epochs. It takes much longer for the model 3 to converge compared to the model 1 and model 2. As the result Shows the overfitting is totally reduced.

### Getting the Sample of the Data Set 
Creating sampling from MNIST digits 5 to 9, using only 100 images per digit

```python
import itertools
X_train_m2 = []
y_train_m2 = []
X_test_m2 = []
y_test_m2 = []
for i in range(5, 10):
    #list1 = [11, 11, 11, 11, 11]
    #list1[i%5] = i
    train_mask = y_train == i
    test_mask = y_test == i
#     train_mask = np.isin(y_train, list1)
#     test_mask = np.isin(y_test, list1)
    X_train_m2.append(X_train[train_mask][:100])
    #X_train.copy()[train_mask][:100, :, :]
    y_train_m2.append(y_train[train_mask][:100])
    X_test_m2.append(X_test[test_mask][:100])
    y_test_m2.append(y_test[test_mask][:100])
    
y_train_m2 = np.concatenate(y_train_m2)
y_test_m2 = np.concatenate(y_test_m2)
X_train_m2 = np.concatenate(X_train_m2)
X_test_m2 = np.concatenate(X_test_m2)
```

```python
X_train_m2, y_train_m2 = unison_shuffled_copies(X_train_m2.copy(), y_train_m2.copy())
X_train_m2, y_train_m2 = reshape(X_train_m2, y_train_m2, num_pixels)
X_test_m2, y_test_m2 = unison_shuffled_copies(X_test_m2, y_test_m2)
X_test_m2, y_test_m2 = reshape(X_test_m2, y_test_m2, num_pixels)
```

## Part 2

1. Create a new DNN that reuses all the pretrained hidden layers of the previous model,
freezes them, and replaces the softmax output layer with a fresh new one.

```python
from keras.models import model_from_json
with open('model.json', 'r') as f:
    new_model = model_from_json(f.read())
new_model.load_weights('model_3_weights.h5')
```

```python
[layer.name for layer in new_model.layers]
```

```python
x = new_model.output
#x = Flatten()(x)
x = Dense(num_classes, activation='softmax')(x)
custom_model = Model(input=new_model.input, output=x)
custom_model.compile(loss='categorical_crossentropy',
                  #optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
              optimizer='adam',
              metrics=['accuracy'])
for layer in custom_model.layers[:10]:
    layer.trainable = False
```

2. Train this new DNN on digits 5 to 9, using only 100 images per digit, and time how long
it takes. Despite this small number of examples, can you achieve high precision?

```python
time_start = time.time()
history3 = custom_model.fit(X_train_m2, y_train_m2, validation_data=(X_test_m2, y_test_m2), epochs=1000, batch_size=100, verbose=2)
time_end = time.time()
print("It took {} to fine tune".format((time_end-time_start % 60)))
scores3 = custom_model.evaluate(X_test_m2, y_test_m2, verbose=0)
#print("{:s}: {:.2f}".format(custom_model.metrics_names[1], score3[1]*100))
print("Baseline Error: %.2f%%" % (100-scores3[1]*100))
```

The accuracy of the test set fluctuates arround 0.5 and 0.6 in the earlier epochs, however it gets higher with the increas of the number of epochs. 5. Now unfreeze the top two hidden layers and continue training: can you get the model
to perform even better?

```python
print("Model: \n{}".format([layer.name for layer in new_model.layers]))
x = new_model.output
#x = Flatten()(x)
x = Dense(num_classes, activation='softmax')(x)
custom_model = Model(input=new_model.input, output=x)
custom_model.compile(loss='categorical_crossentropy',
                  #optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
              optimizer='adam',
              metrics=['accuracy'])
print("The first trainable layer '{}'".format(custom_model.layers[9].name))
for layer in custom_model.layers[:9]:
    layer.trainable = False
for layer in custom_model.layers[:4]:
    layer.trainable = False
```

```python
time_start = time.time()
history3 = custom_model.fit(X_train_m2, y_train_m2, validation_data=(X_test_m2, y_test_m2), epochs=1000, batch_size=100, verbose=2)
time_end = time.time()
print("It took {} to fine tune".format((time_end-time_start % 60)))
scores3 = custom_model.evaluate(X_test_m2, y_test_m2, verbose=0)
#print("{:s}: {:.2f}".format(custom_model.metrics_names[1], score3[1]*100))
print("Baseline Error: %.2f%%" % (100-scores3[1]*100))
```
